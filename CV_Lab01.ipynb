```python
img = data.astronaut()
io.imshow(img)
```

```python
from skimage import color
from skimage.transform import rescale,resize,rotate,downscale_local_mean
image2 = color.rgb2gray(img)
io.imshow(image2)
```

```python
image_rescaled = rescale(image2,5)
io.imshow(image_rescaled)
```

```python
image_resized = resize(image2,(100,100))
io.imshow(image_resized)
```

```python
image_downscale = downscale_local_mean(image2,(3,3))
io.imshow(image_downscale)
```

```python
image_rotate = rotate(image2,45)
io.imshow(image_rotate)
```
