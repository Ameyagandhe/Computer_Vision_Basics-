```
import numpy as np
import pandas as pd
import cv2 as cv2
from skimage import io
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from PIL import Image
#chttps://scikit-image.org/docs/dev/api/skimage.data.html
```

```
from skimage import data
coins = data.coins()
io.imshow(coins)
```

```
kernal = np.ones([5,5])/25
dst = cv2.filter2D(coins,-1,kernal)
io.imshow(dst)
plt.title("Homogenous filter using filter2D function")
```

```
plt.subplot(1,2,1),io.imshow(coins),plt.title("orignal")
plt.subplot(1,2,2),io.imshow(dst),plt.title("Homogenous filter using filter2D function ")
```

```
blur = cv2.blur(coins,(5,5))
io.imshow(blur)
plt.title("using blur function")
```

Gaussian Blur

```
G_blur = cv2.GaussianBlur(coins,(5,5),1)
io.imshow(G_blur)
plt.title("using Gaussian blur function")
```

```
plt.subplot(1,2,1),io.imshow(G_blur),plt.title("gaussian blur")
plt.subplot(1,2,2),io.imshow(dst),plt.title("Homogenous filter using filter2D function ")
```

```
bilateral = cv2.bilateralFilter(coins,5,75,5,75)
io.imshow(bilateral)
```

```
plt.subplot(1,3,1),io.imshow(coins),plt.title("orignal")
plt.subplot(1,3,2),io.imshow(dst),plt.title("Homogenous filter using filter2D function ")
plt.subplot(1,3,3),io.imshow(bilateral),plt.title("Bilateral ")
```

```
M_blur = cv2.medianBlur(coins,5)
io.imshow(M_blur)
```

```
Lap = cv2.Laplacian(coins,cv2.CV_64F)
Lap = np.uint8(np.absolute(Lap))
io.imshow(Lap)
```

```
chess = data.checkerboard()
# io.imshow(chess)
sobely = cv2.Sobel(chess,cv2.CV_64F,1,0)
sobelx = cv2.Sobel(chess,cv2.CV_64F,0,1)
sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))
io.imshow(sobelx)
```

```
io.imshow(sobely)
```

```
sobelcombined = cv2.bitwise_or(sobelx,sobely)
io.imshow(sobelcombined)
```

```
sobelCoiny = cv2.Sobel(coins,cv2.CV_64F,1,0)
sobelCoinx = cv2.Sobel(coins,cv2.CV_64F,0,1)
sobelCoinx = np.uint8(np.absolute(sobelCoinx))
sobelCoiny = np.uint8(np.absolute(sobelCoiny))
sobelCoinCombinedCoin = cv2.bitwise_or(sobelCoinx,sobelCoiny)
io.imshow(sobelCoinCombinedCoin)
```

```
plt.subplot(1,2,1)
io.imshow(Lap)
plt.title("Laplacian")
plt.subplot(1,2,2)
io.imshow(sobelCoinCombinedCoin)
plt.title("Sobel combined")
```
